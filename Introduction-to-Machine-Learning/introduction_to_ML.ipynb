# Introduction to Machine Learning

<b>"A computer program is said to learn from experience ùê∏ with respect to some task ùëá and some performance measure ùëÉ, if its performance on ùëá, as measured by ùëÉ, improves with experience ùê∏."</b> ‚Äì Tom Mitchell, Carnegie Mellon University

experience E -> data

task T -> loss function

performans measure P -> metric(s)

We have data, function that we optimize, and metrics attached to it. The optimizer train an ML algorithm (i.e. the loss function is minimized), and thus, some metrics improve.

<b>The algorith in computer science</b> -> the metrics do not improve regardless how many data we input into it.

<b>ML algorithm</b> -> an empty model

## Types of ML algorithms

![Alt text](ml_algorithm.png)

<b>Supervised learning</b> -> train the program on labelled data and expect the model to make predictions on new data. Examples: regression, classification.

<b>Unsupervised learning</b> -> the program finds patterns in data. We do not have an expected result, but we have some expectations to the data i.e we have a hypothesis. Examples: clustering analysis, dimensionality reduction.

<b>Reinforcement learning</b> -> agents, an adavtive algorithm. A combination of supervised and unsupervised learning, the program learns continuously. Examples: learning something by observing other things/actions in the real world; learning to drive a car.

## Algorithms by Task

<b>Statistical algorithms</b>

<b>Regression</b> ‚Äì> predicting a continuous variable

<b>Classification</b> ‚Äì> predicting class labels

<b>Clustering</b> ‚Äì> finding compact groups of data points

<b>Dimensionality reduction</b> ‚Äì> simplifying the input data

<b>Recommendation</b> ‚Äì> suggest items for users

<b>Optimization</b> ‚Äì> minimize / maximize a target function

<b>Testing and improvement algorithms</b> ‚Äì> helper algorithms to select, fine-tune and optimize other ML algorithms

and so on...
